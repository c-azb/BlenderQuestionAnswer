```python
import os
import sys
from dotenv import load_dotenv

APP_DIR = os.path.abspath('../BlenderQuestionAnswer_app')
VECTOR_DATABASES_DIR = os.path.join(APP_DIR,'Blender_Agentic_RAG/VectorDataBases')
sys.path.append(APP_DIR)
load_dotenv('C:/env/.env')
```

```python
from  Blender_Agentic_RAG.graph_ai.agentic_main import AgenticGraph,LLMSources,EmbeddingsSource,VectorDBSource,SearchState,MemoryType
```

```python
agentic_graph = AgenticGraph(blender_version = '4.5',
                            main_model = 'llama3.2:3b',#llama3.1:8b
                            source = LLMSources.OLLAMA,
                            deterministic_model = None,
                            recom_model = None,
                            creative_model = None,
                            include_tavily_answer=False,
                            max_tavily_results=3,
                            n_results_vdb = 4,
                            search_variation_vdb = 0.9,
                            vdbs_path = VECTOR_DATABASES_DIR,
                            embeddings_source = EmbeddingsSource.OLLAMA,
                            vectordb_source = VectorDBSource.FAISS,
                            embedding_name = 'nomic-embed-text')
```

```python
agentic_graph.graph
```

```python
response = agentic_graph.invoke(
                        question = 'How to extrude a face?', 
                        session_id = 0, 
                        search_state = SearchState.FIRST_SEARCH, 
                        allow_external_search = False,
                        memory_type = MemoryType.TEMPORARY, 
                        ai_memory = None,
                        stream_mode = 'values' #updates, values
                        )
```
